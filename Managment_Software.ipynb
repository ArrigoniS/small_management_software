### Funzioni accessorie e di controllo

```python
# Function for access and control
def numero_intero(numero):
    '''
    Function that checks whether the entered value is a number and can be converted into an integer.
    '''
    while not type(numero) is int:
        try:
            numero = int(numero)
        except:
            print("\nThe number insert is not valid!")
            numero = input("\nInsert again the quantity")
    return numero

def numero_float(numero):
    '''
    Function that checks whether the entered value is a number and can be converted into a float.
    '''
    while not type(numero) is float:
        try:
            numero = float(numero)
        except:
            print("\nThe number insert is not valid!")
            numero = input("\nInsert again the quantity")
    return numero

def continuare_check(continuare_stringa):
    '''
    Function that checks whether the entered value is yes or no.
    '''
    while continuare_stringa not in ["yes","no"]:
        print("\nValue not valid. Insert only yes or no.")
        continuare_stringa = input("\nInsert a command? yes/no   ").casefold()
    return continuare_stringa

def creazione_stringa(prodotto):
    '''
    Auxiliary function for printing the inventory on screen. It allows aligning the different rows being printed.
    prodotto (str): name of the product to be printed.
    '''
    len_nome = len(" Product name ")
    len_quantita = len(" Quantity in stocl ")
    len_prezzo_acq = len(" Acquiring price ")
    len_prezzo_ven = len(" Selling price")
    len_tas = len(" Tax applied")

    lunghezza1 = len(str(prodotto))
    lunghezza_spazi1 = math.floor((len_nome - lunghezza1)/2)
    if len(str(prodotto))%2 == 1:
        spazi1a = " "*(lunghezza_spazi1 - 1)
        spazi1b = " "*(lunghezza_spazi1 + 1)
    else:
        spazi1b = " "*lunghezza_spazi1
        spazi1a =  spazi1b
    nome = spazi1a + str(prodotto) + spazi1b

    lunghezza2 = len(str(prodotti_dict[prodotto][0]))
    lunghezza_spazi2 = math.floor((len_quantita - lunghezza2)/2)
    if len(str(prodotto))%2 == 1:
        spazi2a = " "*(lunghezza_spazi2 - 1)
        spazi2b = " "*(lunghezza_spazi2 + 1)
    else:
        spazi2b = " "*lunghezza_spazi2
        spazi2a =  spazi2b
    quantita = spazi2a + str(prodotti_dict[prodotto][0]) + spazi2b

    lunghezza3 = len(str(prodotti_dict[prodotto][1]))
    lunghezza_spazi3 = math.floor((len_prezzo_acq - lunghezza3)/2) - 1
    if len(str(prodotto))%2 == 1:
        spazi3a = " "*(lunghezza_spazi3 - 1)
        spazi3b = " "*(lunghezza_spazi3 + 1)
    else:
        spazi3b = " "*lunghezza_spazi3
        spazi3a =  spazi3b
    prezzo_acq = spazi3a + str(prodotti_dict[prodotto][1]) + "€" + spazi3b

    lunghezza4 = len(str(prodotti_dict[prodotto][2]))
    lunghezza_spazi4 = math.floor((len_prezzo_ven - lunghezza4)/2) - 1
    if len(str(prodotto))%2 == 1:
        spazi4a = " "*(lunghezza_spazi4 - 1)
        spazi4b = " "*(lunghezza_spazi4 + 1)
    else:
        spazi4b = " "*lunghezza_spazi4
        spazi4a =  spazi4b
    prezzo_ven = spazi4a + str(prodotti_dict[prodotto][2]) + "€" + spazi4b

    lunghezza5 = len(str(prodotti_dict[prodotto][3]))
    lunghezza_spazi5 = math.floor((len_tas - lunghezza5)/2) - 1
    if len(str(prodotto))%2 == 1:
        spazi5a = " "*(lunghezza_spazi5 - 2)
        spazi5b = " "*(lunghezza_spazi5)
    else:
        spazi5b = " "*lunghezza_spazi5
        spazi5a =  spazi5b
    tass = spazi5a + str(prodotti_dict[prodotto][3]) + "%" + spazi5b

    indici = []
    stringa_base = "\nProduct name | Quantity in stock | Acquiring price | Selling price | Tax Applied"
    for i, c in enumerate(stringa_base):
            if c == "|":
                indici.append(i)

    stringa = f"{nome}  {quantita}  {prezzo_acq}   {prezzo_ven}   {tass}"
    stringa = stringa[:indici[0]-1] + "|" + stringa[indici[0]:indici[1]-1]  + "|" + stringa[indici[1]:indici[2]-1] + "|" + stringa[indici[2]:indici[3]-1] + "|" + stringa[indici[3]+1:]
    return stringa

def creazione_stringa_profitti(prodotto, profitto_lordo, profitto_netto):
    '''
    Auxiliary function for printing profits on screen. It allows aligning the different rows being printed.
    prodotto (str): name of the product to be printed.
    profitto_lordo (float): gross profit calculated for the specific product.
    profitto_netto (float): net profit calculated for the specific product.
    '''
    len_nome = len("Product name ")
    len_lordo = len(" Gross profit ")
    len_netto = len(" Net profit")

    lunghezza1 = len(str(prodotto))
    lunghezza_spazi1 = math.floor((len_nome - lunghezza1)/2)
    if len(str(prodotto))%2 == 1:
        spazi1a = " "*(lunghezza_spazi1 + 1)
        spazi1b = " "*(lunghezza_spazi1 )
    else:
        spazi1b = " "*lunghezza_spazi1
        spazi1a =  spazi1b
    nome = spazi1a + str(prodotto) + spazi1b

    lunghezza2 = len(str(profitto_lordo))
    lunghezza_spazi2 = math.floor((len_lordo - lunghezza2)/2)
    if len(str(prodotto))%2 == 2:
        spazi2a = " "*(lunghezza_spazi2 + 1)
        spazi2b = " "*(lunghezza_spazi2)
    else:
        spazi2b = " "*lunghezza_spazi2
        spazi2a =  spazi2b
    profitto_lordo = spazi2a + str(profitto_lordo) + "€" + spazi2b

    lunghezza3 = len(str(profitto_netto))
    lunghezza_spazi3 = math.floor((len_netto - lunghezza3)/2) - 1
    if len(str(prodotto))%2 == 2:
        spazi3a = " "*(lunghezza_spazi3 + 1)
        spazi3b = " "*(lunghezza_spazi3)
    else:
        spazi3b = " "*lunghezza_spazi3
        spazi3a =  spazi3b
    profitto_netto = spazi3a + str(profitto_netto) + "€" + spazi3b

    indici = []
    stringa_base = "\nProduct Name | Gross Profit | Net Profit"
    for i, c in enumerate(stringa_base):
            if c == "|":
                indici.append(i)

    stringa = f"{nome}  {profitto_lordo}  {profitto_netto}"
    stringa = stringa[:indici[0]] + "|" + stringa[indici[0]+1:indici[1]]  + "|" + stringa[indici[1]+1:]
    return stringa
```

### Functionality of the managment software

1. aggiungi_prodotto: allows adding a product to the inventory.
2. elenca_prodotti: allows listing the products currently in the inventory.
3. elenco_profitti: allows listing the standard gross and net profit for a specific product or for all products in the inventory.
4. scontrino: allows printing the receipt after a sale.
5. registrazione_vendita: allows recording a sale in the cash register (a CSV file that is generated when the software is closed).   


```python
# Functionality function
def aggiungi_prodotto(info_prodotto, prodotti_dict):
    '''
    Function used to add a product to the warehouse inventory.
    info_prodotto: list of information required for registration [name, quantity, purchase price, selling price, applied tax]
    prodotti_dict: dictionary of the current inventory
    '''
    if len(info_prodotto) == 5:
        prodotti_dict[info_prodotto[0]] = info_prodotto[1:]
    else:
        print("\nAction Denied. The information inserted are wrong.\n")
    return prodotti_dict


def elenca_prodotti(prodotti_dict):
    '''
    Function that prints on screen all the information of all products present in the warehouse.
    prodotti_dict: dictionary of the current inventory
    '''
    lista_prodotti = list(prodotti_dict.keys())
    print("\n#   | Product Name | Quantity in stock | Acquiring price per piece | Selling Price | Tax Applied")
    j = 1
    for i in lista_prodotti:
        stringa = creazione_stringa(i)
        if len(str(j)) == 1:
            num = str(j) + "  "
        if len(str(j)) == 2:
            num = str(j) + " "
        print(f"{num} | {stringa}")
        j += 1

def elenco_profitti(prodotto_input, prodotti_dict):
    '''
    Function used to print on screen the profits (gross and net) of a specific product or the entire inventory.
    prodotto_input (str): name of the product or "all" if you want to print for the entire inventory.
    prodotti_dict: dictionary of the current inventory
    '''
    if prodotto_input == "all":
        prodotti_keys = list(prodotti_dict.keys())
        print("\nProduct Name | Gross Profit | Net Profit")
        for prodotto in prodotti_keys:
            profitto_lordo = round(prodotti_dict[prodotto][2] - prodotti_dict[prodotto][1], 2)
            profitto_netto = round(profitto_lordo - (profitto_lordo*prodotti_dict[prodotto][3])/100, 2)
            stringa = creazione_stringa_profitti(prodotto, profitto_lordo, profitto_netto)
            print(f"{stringa}")
    else:
        print("\nProduct Name | Gross Profit | Net Profit")
        profitto_lordo = round(prodotti_dict[prodotto_input][2] - prodotti_dict[prodotto_input][1], 2)
        profitto_netto = round(profitto_lordo - (profitto_lordo*prodotti_dict[prodotto_input][3])/100, 2)
        stringa = creazione_stringa_profitti(prodotto_input, profitto_lordo, profitto_netto)
        print(f"{stringa}")

def scontrino(info_vendita, prodotti_dict):
    '''
    Function used to print the receipt on screen and calculate the total amount due.
    info_vendita: information required to calculate the total and print the receipt; a list of lists in the format [product name, quantity purchased]
    prodotti_dict: dictionary of the current inventory
    '''
    lista_totali = []
    for i in range(len(info_vendita)):
        print(f"{info_vendita[i][0]}     {prodotti_dict[info_vendita[i][0]][2]}€ x {info_vendita[i][1]}     {round(prodotti_dict[info_vendita[i][0]][2]*info_vendita[i][1],2)}€\n")
        lista_totali.append(prodotti_dict[info_vendita[i][0]][2]*info_vendita[i][1])
    print("-"*18)
    print(f"Total     {round(sum(lista_totali),2)} €")
    return round(sum(lista_totali),2)



def registrazione_vendita(spesa, tipo_pagamento, totale_pagato):
    '''
    Function that records the most recent sale and saves it in the cash register.
    spesa: information required to calculate the total gross and net profit; a list of lists in the format [product name, quantity purchased]
    tipo_pagamento (str): type of payment made by the customer (e.g., card or cash)
    totale_pagato (float): total amount of the purchase
    '''
    totale_profitti_lordo = []
    totale_profitti_netto = []
    for prodotto, numero in spesa:
        profitto_lordo = (prodotti_dict[prodotto][2] - prodotti_dict[prodotto][1]) * numero
        profitto_netto = (profitto_lordo - (profitto_lordo*prodotti_dict[prodotto][3])/100) * numero
        totale_profitti_lordo.append(round(profitto_lordo,2))
        totale_profitti_netto.append(round(profitto_netto,2))

    registro_vendite["Data"].append(str(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")))
    registro_vendite["Payment Type"].append(tipo_pagamento)
    registro_vendite["Total payied"].append(str(totale_pagato)+"€")
    registro_vendite["Gross Profit"].append(str(sum(totale_profitti_lordo))+"€")
    registro_vendite["Net Profit"].append(str(sum(totale_profitti_netto))+"€")
```

### Funzioni che permettono l'apertura e la chiusura del software gestionale

```python
# Funzioni che permettono l'apertura e la chiusura del software di cassa

def apertura_software():
    '''
    Function that launches the cash register software by searching for the CSV files used as a database, if present. Returns the inventory and the cash register in dictionary format.
    inventario = {product_name_1: [stock quantity, purchase price, selling price, applied tax],,
                  product_name_2: [stock quantity, purchase price, selling price, applied tax]}
    registro di cassa = "Date": [purchase_date_1, purchase_date_2, ...],
                        "Payment Type": [payment_type_1, payment_type_2, ...],
                        "Total Paid": [total_1, total_2, ...],
                        "Gross Profit": [gross_profit_1, gross_profit_2, ...],
                        "Net Profit": [net_profit_1, net_profit_2, ...]
    '''
    inventario = {}
    try:
        with open("inventario.csv") as file:
            csv_reader = csv.reader(file)
            lista_reader = list(csv_reader)
            for riga in lista_reader[1:]:
                riga = riga[0].split(";")
                riga[1] = int(riga[1])
                riga[2] = float(riga[2])
                riga[3] = float(riga[3])
                riga[4] = float(riga[4])
                inventario[riga[0].casefold()] = riga[1:]
    except:
        print("Warning: Empty inventory. If an error occurs, reload and check that the inventory CSV file is in the correct folder.")


    registro_cassa = {"Data": [],
                        "Payment Type" : [],
                        "Total Payied" : [],
                        "Gross Profit": [],
                        "Net Profit" : []}
    try:
        with open("registro_cassa.csv") as file:
            csv_reader = csv.reader(file)
            lista_reader = list(csv_reader)
            for riga in lista_reader[1:]:
                riga = riga[0].split(";")
                registro_cassa["Data"].append(riga[0])
                registro_cassa["Payment Type"].append(riga[1])
                registro_cassa["Total Payied"].append(riga[2])
                registro_cassa["Gross Profit"].append(riga[2])
                registro_cassa["Net Profit"].append(riga[4])
    except:
        print("Warning: Sales register is empty. If an error occurs, reload and check that the cash register CSV file is in the correct folder.")

    return inventario, registro_cassa

def chiusura_software(inventario, registro_cassa):
    '''
    Function that allows saving the current inventory and the cash register in CSV format.
    '''
    salvataggio_dati = [["Product Name;Quantity;Purchase Price; Selling Price; Percentage of Tax"]]
    lista_prodotti = list(inventario.values())
    i = 0
    for chiave in list(inventario.keys()):
        quantita = str(lista_prodotti[i][0])
        prezzo_acq = str(lista_prodotti[i][1])
        prezzo_ven = str(lista_prodotti[i][2])
        tass = str(lista_prodotti[i][3])

        riga = [chiave + ";" + quantita + ";" + prezzo_acq + ";" + prezzo_ven + ";" + tass]
        salvataggio_dati.append(riga)

        i += 1

    with open("inventario.csv", "w", newline = '') as file:
        writer = csv.writer(file)
        for riga in salvataggio_dati:
            writer.writerow(riga)


    salvataggio_registro_cassa = [["Data;Payment Type;Total Payied;Gross Profit;Net Profit"]]
    for i in range(len(registro_cassa["Data"])):
        riga = [str(registro_cassa["Data"][i]) + ";" + str(registro_cassa["Payment Type"][i]) + ";" + str(registro_cassa["TTotal Payied"][i]) + ";" + str(registro_cassa["Gross Profit"][i]) + ";" + str(registro_cassa["Net Profit"][i])]
        salvataggio_registro_cassa.append(riga)

    with open("registro_cassa.csv", "w", newline = '') as file:
            writer = csv.writer(file)
            for riga in salvataggio_registro_cassa:
                writer.writerow(riga)
```

## Managment Software

#### **Command**:
- **"Help"**: Function that provides a list of the available commands.
- **"add"**: allows inserting new ordered products and adding them to the warehouse inventory. (Note: before executing, it is necessary to know the tax applied to the item)
- **"list"**: allows listing the products in the warehouse along with their details (quantity, selling and purchase price, and tax)
- **"sale"**: allows performing a sale, showing the final receipt and adding the information to the cash register
- **"profits"**: allows viewing the profit per item for each specific product, or by typing "all" it is possible to view profits for the entire inventory
- **"change price"**:allows changing the selling price of a specific product
- **"change tax"**:allows changing the applied tax for a specific product
- **"delete"**: allows deleting a specific product from the inventory
- By entering the name of a product present in the warehouse, it is possible to view all the system-stored information related to that product



```python
# Inizializzazione del software di cassa
prodotti_dict,registro_vendite = apertura_software()
lista_prodotti = list(prodotti_dict.keys())
progressivo = 0

continuare = input("Start the software? yes/no   ").casefold()
continuare = continuare_check(continuare)


while continuare == "yes":
    print("_"*50)
    comando = input("\nWhich command do you want to do?").casefold()
    # Ciclo if che permette di scegliere la funzionalità ricercata

    if comando == "add":
        # Aggiungire un prodotto all'inventario
        print("_"*50)
        nome_prodotto = input("\nInsert the product name   ").casefold()
        quantità = input("\nInsert the quantity purchased.   ")
        quantità = numero_intero(quantità)
        prezzo_acquisto = input("\nHow much it was per piece?   ")
        prezzo_acquisto = numero_float(prezzo_acquisto)
        prezzo_vendita = input("\nHow much would it be the selling price?   ")
        prezzo_vendita = numero_float(prezzo_vendita)
        perc_tassazione = input("\nWhat's going to be the tax applied? Insert only the number.   ")
        perc_tassazione = numero_float(perc_tassazione)
        prodotti_dict = aggiungi_prodotto([nome_prodotto, quantità, prezzo_acquisto, prezzo_vendita, perc_tassazione], prodotti_dict)
        lista_prodotti = list(prodotti_dict.keys())
    elif comando == "list":
        # Elencare l'inventario corrente
        print("_"*50)
        elenca_prodotti(prodotti_dict)
    elif comando in lista_prodotti:
        # Elencare i dati di uno specifico prodotto
        print("_"*50)
        stringa = creazione_stringa(comando)
        print("\nProduct Name | Quantity in stock | Purchased price | Selling price | Tax applied")
        print(f"{stringa}\n")
    elif comando == "sale":

        # Effettuare una vendita (registrazione, scontrino e calcolo del totale e dei profitti)
        print("_"*50)
        vendita = []
        altro = "yes"
        while altro == "yes":

            # Creazione delle lista della spesa (controllo su quantità e nome del prodotto)
            nome_prodotto = input("\nWhich product are you selling?   ").casefold()
            while nome_prodotto not in lista_prodotti:
                nome_prodotto = input("\nProduct not present in stock. Re-insert the product name.   ").casefold()
            quantità = input("How much product are you selling?   ")
            quantità = numero_intero(quantità)
            while quantità > prodotti_dict[nome_prodotto][0]:
                if prodotti_dict[nome_prodotto][0] == 0:
                    nome_prodotto = input("\nIThere is not enough stock. Choose another item to sell.   ").casefold()
                    quantità = int(input("How much product are you selling?   "))
                    quantità = numero_intero(quantità)
                else:
                    print(f"Not enough in stock. Max quantity: {prodotti_dict[nome_prodotto][0]}")
                    quantità = int(input("Re-insert the quantity. If not sell anymore put 0   "))
                    quantità = numero_intero(quantità)

            prodotti_dict[nome_prodotto][0] = prodotti_dict[nome_prodotto][0] - quantità
            vendita.append([nome_prodotto, quantità])
            print(f"\n{nome_prodotto} {prodotti_dict[nome_prodotto][2]}€ x {quantità}")
            altro = input("Other item? yes/no   ").casefold()
        print(" ")

        #Stampa dello scontrino
        print("*"*40)
        nome_market = " "*12 + "BioMarket s.a.s."
        indirizzo = " "*5 + "Address: Via Tan 6, Milano"
        partita_iva = " "*13 + "84409660853"
        print(f"{nome_market} \n{indirizzo} \n {partita_iva}\n")

        totale = scontrino(vendita,prodotti_dict)

        print(f"\n{datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}")
        progressivo += 1
        print(f"daily progress number: {progressivo}")
        print("*"*40)

        # Registrazione di cassa
        tipologia_pagamento = input("\nCash or card?")
        registrazione_vendita(vendita, tipologia_pagamento, totale)

        print("\nSale done successfully.")
    elif comando == "profits":
        # Elenco dei profitti per specifico prodotto o per tutti i prodotti in inventario
        print("_"*50)
        nome_prodotto = input("\nSpecify a particular product or type “all” to view the profits.").casefold()
        while (nome_prodotto not in lista_prodotti) and (nome_prodotto != "all"):
                nome_prodotto = input("\nThe product does not exist in the warehouse, please re-enter the name.   ").casefold()
        elenco_profitti(nome_prodotto, prodotti_dict)
    elif comando == "change price":
        # Comando che permette di cambiare il prezzo di vendita di uno specifico prodotto
        nome_prodotto = input("\nFor which product?").casefold()
        while (nome_prodotto not in lista_prodotti):
            nome_prodotto = input("\nThe product does not exist in the warehouse, please re-enter the name. ").casefold()
        nuovo_prezzo = input("\nWhat's the new selling price?")
        nuovo_prezzo = numero_float(nuovo_prezzo)
        prodotti_dict[nome_prodotto][2] = nuovo_prezzo
    elif comando == "cange tax":
        # Comando che permette di cambiare il prezzo di vendita di uno specifico prodotto
        nome_prodotto = input("\nFor which product?").casefold()
        while (nome_prodotto not in lista_prodotti):
            nome_prodotto = input("\nThe product does not exist in the warehouse, please re-enter the name ").casefold()
        nuovo_tassazione = input("\nWhat's the new tax?")
        nuovo_tassazione = numero_float(nuovo_tassazione)
        prodotti_dict[nome_prodotto][3] = nuovo_tassazione
    elif comando == "delete":
        # Comando che permette di rimuovere uno specifico prodotto dall'inventario
        nome_prodotto = input("\nWhich product?").casefold()
        while (nome_prodotto not in lista_prodotti):
            nome_prodotto = input("\nThe product does not exist in the warehouse, please re-enter the name ").casefold()
        del prodotti_dict[nome_prodotto]

    elif comando == "Help":
        # Comando che permette di avere un elenco delle funzionalità disponibili in caso di richiesta
        print("_"*50)
        print("\nList of command: \n")
        print("- add: allows inserting new ordered products and adding them to the warehouse inventory (before executing, it is necessary to know the tax applied to the item).")
        print("- list: allows listing the products in the warehouse and their respective quantities.")
        print("- By entering the name of a product present in the warehouse, it is possible to view all the information stored in the system related to that product.")
        print("- sale: allows performing a sale, displaying the final receipt, and adding the information to the cash register.")
        print("- profits: allows viewing the profit per item for each specific product, or by typing \"all\" it is possible to view profits for the entire inventory.")
        print("- change price: allows changing the selling price of a specific product.")
        print("- change tax: allows changing the applied tax for a specific product.")
        print("- delete: allows removing a specific product from the inventory.")
    else:
        # Comando che permette di avere un elenco delle funzionalità disponibili in caso di errore nella compilazione
        print("_"*50)
        print("\nInvalid command\n")
        print("\nList of command: \n")
        print("- add: allows inserting new ordered products and adding them to the warehouse inventory (before executing, it is necessary to know the tax applied to the item).")
        print("- list: allows listing the products in the warehouse and their respective quantities.")
        print("- By entering the name of a product present in the warehouse, it is possible to view all the information stored in the system related to that product.")
        print("- sale: allows performing a sale, displaying the final receipt, and adding the information to the cash register.")
        print("- profits: allows viewing the profit per item for each specific product, or by typing \"all\" it is possible to view profits for the entire inventory.")
        print("- change price: allows changing the selling price of a specific product.")
        print("- change tax: allows changing the applied tax for a specific product.")
        print("- delete: allows removing a specific product from the inventory.")

    print("\n")
    print("_"*50)
    continuare = input("\nYou want to do another command? yes/no   ").casefold()
    continuare = continuare_check(continuare)

# Chiusura del software e salvataggio dei dati necessari
print("_"*100)
print("\nClosing. Saving database.\n")

chiusura_software(prodotti_dict, registro_vendite)

print("Closed. \nNow it's possible to visulize the data in the two csv file.\nGood Bye!.")
```
